```python
from selenium import webdriver
from time import sleep

# defining new variable passing two parameters
# writer = csv.writer(open(parameters.file_name, 'wb'))

# # writerow() method to the write to the file object
# writer.writerow(['Name', 'Job Title', 'Company', 'College', 'Location', 'URL'])

# specifies the path to the chromedriver.exe
driver = webdriver.Chrome()

# driver.get method() will navigate to a page given by the URL address
driver.get('https://www.linkedin.com/login')

# locate email form by_class_name
username = driver.find_element_by_id('username')

# send_keys() to simulate key strokes
username.send_keys('rubyyang0615@gmail.com')

# sleep for 0.5 seconds
sleep(0.5)

# locate password form by_class_name
password = driver.find_element_by_id('password')

# send_keys() to simulate key strokes
password.send_keys('lina311')
sleep(0.5)

# locate submit button by_xpath
sign_in_button = driver.find_element_by_class_name('btn__primary--large')

# .click() to mimic button click
sign_in_button.click()
sleep(5)

driver.get('https://www.linkedin.com/groups/10463057/members/')
sleep(35)

tmp = driver.find_elements_by_css_selector('.artdeco-entity-lockup__content > a')

# store link to list
profiles = []
for item in tmp:
    profiles.append(item.get_attribute('href'))

print(profiles)

pages = []
for url in profiles:
    driver.get(url)
    sleep(20)
    pages.append(driver.page_source)
    sleep(5)

driver.quit()
```

```python
import csv
with open('profile_pages_190824.csv', 'w', newline='') as file:
    wr = csv.writer(file, quoting=csv.QUOTE_ALL)
    wr.writerow(pages)
```

```python
import pandas as pd

profile_data = []
tmp = []
for page in pages:
    page = BeautifulSoup(page)
    name = page.find('li', {'class': ['inline', 't-24']}).get_text()
    if name:
        name = name.strip()
    print(name)
    location = page.select('li.t-normal.inline-block.t-16.t-black')[0].get_text()
    if location:
        location = location.strip()
    print(location)
    
#     work_exp = page.xpath('//section[@id="experience-section"]').extract_first()
    job_titles = page.find('section', {'id': 'experience-section'}).findAll('h3', {'class': 't-bold'})
    for job_title in job_titles:
        job_title = job_title.getText()
        if job_title:
            job_title = job_title.strip()
        print(job_title)
        tmp.append(job_title)
    job_titles = tmp
    tmp = []
    companies = page.find('section', {'id': 'experience-section'}).findAll('span', {'class': 'pv-entity__secondary-title'})
    for company in companies:
        company = company.getText()
        if company:
            company = company.strip()
        print(company)
        tmp.append(company)
    companies = tmp
    tmp = []
    dateranges = page.find('section', {'id': 'experience-section'}).findAll('h4', {'class': 'pv-entity__date-range'})
    for daterange in dateranges:
        daterange = daterange.getText()
        if daterange:
            daterange = daterange.strip()
        print(daterange)
        tmp.append(daterange)
    dateranges = tmp
    tmp = []
    edu_school = page.select('section#education-section h3')[0].get_text()
    if edu_school:
        edu_school = edu_school.strip()
    print(edu_school)
    
    skills = page.findAll('span', {'class': 'pv-skill-category-entity__name-text'})
    for skill in skills:
        skill = skill.getText()
        if skill:
            skill = skill.strip()
        print(skill)
        tmp.append(skill)
    skills = tmp
    tmp = []
    skill_endorsements = page.findAll('span', {'class': 'pv-skill-category-entity__endorsement-count'})
    for skill_endorsement in skill_endorsements:
        skill_endorsement = skill_endorsement.getText()
        if skill_endorsement:
            skill_endorsement = int(skill_endorsement.strip())
        print(skill_endorsement)
        tmp.append(skill_endorsement)
    skill_endorsements = tmp
    tmp = []
    profile_data.append([name, location, job_titles, companies, dateranges, edu_school, skills, skill_endorsements])
    
df_profile = pd.DataFrame(profile_data, columns=['Name', 'Location', 'Job_Titles', 'Companies', 'Date_Ranges', 'Edu_School', 'Skills', 'Skill_Endorsements'])

df_profile
```
